# 10.6 관계 시각화

- 산점도를 그릴 때, 극단치로 인해 주요 분포 구간이 압축되어 시각화 효율이 떨어지기 때문에 극단치는 제거하는 것이 좋다.
- 데이터가 너무 많아 점들이 겹쳐 보기 힘들 경우, 각 점에 투명도를 주어 밀도를 함께 표현하거나 빈도에 따라 색상을 다르게 표현해주면 좋다.
- 산점도는 두 개의 변수 간 관계만 표현할 수 있다.
- **버블 차트**는 세 가지 요소의 상관 관계를 표현할 수 있다.
- 버블 차트를 해석할 때, 원의 지름이 아닌 면적을 통해 크기를 판단해야 하는 것 주의! (지름이 두 배 크면 실제 크기는 네 배가 큰 것이므로) 

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
# 스타트업 기업들의 R&D 비용, 관리 비용, 마케팅 비용 등 수익 데이터
df = pd.read_csv('datasets/50_Startups.csv')
df.head()
```

## 기본 산점도 시각화

```python
plt.scatter(df['R&D Spend'], df['Profit'], s=50, alpha=0.4)
plt.show()

# s: 점 크기
# alpha: 투명도
```

```python
# 산점도에 회귀선 추가
ax = sns.lmplot(x='R&D Spend', y='Profit', data=df)
```

```python
# 네 가지 요소의 정보를 포함한 산점도 시각화 (버블 차트)

plt.scatter(df['R&D Spend'], df['Profit'], s=df['Marketing Spend']*0.001, 
           c=df['Administration'], alpha=0.5, cmap='Spectral')

plt.colorbar()
plt.show()


# 버블 크기(s): 마케팅 비용
# 버블 색상(c): 관리 비용

# 마케팅 비용과 관리 비용도 수익과 어느 정도 상관 관계가 있는 것으로 보인다.
```
